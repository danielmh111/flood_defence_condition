```python
date_cols = ["last_inspection_date", "next_inspection_date", "asset_start_date", "current_sop_date"]
date_pattern = r"^\d{2}/\d{2}/\d{4}$"

date_audit = []
for col in date_cols:
    non_null = gdf[col].dropna()
    # also exclude whitespace-only
    non_empty = non_null[non_null.str.strip() != ""]
    matches = non_empty.str.match(date_pattern)
    non_matching = non_empty[~matches]
    date_audit.append({
        "column": col,
        "non_empty_count": len(non_empty),
        "matches_DD/MM/YYYY": matches.sum(),
        "exceptions": len(non_matching),
        "exception_samples": non_matching.head(5).tolist() if len(non_matching) > 0 else [],
    })

date_audit_df = pd.DataFrame(date_audit).set_index("column")
date_audit_df
```

coverage is unfortunate. Only 25% of the assets actually have a condition grade. This gives a much smaller dataset than i was hoping for, but actually it could bolster the explaination for the need for this research.

I need to check if the whitespace values occur with any pattern. For example, i wonder if they are all for natural highground features that aren't actually inspectable and gradable. If so, they ought to be excluded from the dataset anyway because its probably invalid to say we can predict wear and tear on a sleuce channel and the errosion to a natural hill in the same way. 

Worst case scenario, we end up removing lots of natural highground AND lots of ungraded defenses. 

Another possiblity is that the ones with missing condition grades are the same as the ones with no last inspection date, meaning they have never been inspected. 

Three hypotheses:
1. Whitespace grades cluster in specific asset subtypes (e.g. Natural High Ground)
2. Whitespace grades coincide with missing `last_inspection_date` (never inspected)
3. Some combination of both

```python
# Tag each row with its condition status
gdf["_cond_status"] = "graded"
gdf.loc[is_null, "_cond_status"] = "null"
gdf.loc[is_whitespace, "_cond_status"] = "whitespace"
gdf.loc[stripped == "0", "_cond_status"] = "grade_0"
```

### Hypothesis 1: Whitespace grades by asset sub-type

```python
# Cross-tab: asset_sub_type vs condition status
ct = pd.crosstab(gdf["asset_sub_type"], gdf["_cond_status"])
ct["total"] = ct.sum(axis=1)
ct["whitespace_pct"] = (ct["whitespace"] / ct["total"] * 100).round(1)
ct = ct.sort_values("total", ascending=False)
ct
```

```python
# Where do the whitespace rows actually come from?
ws = gdf[gdf["_cond_status"] == "whitespace"]
ws_by_type = ws["asset_sub_type"].value_counts()
ws_by_type_df = ws_by_type.to_frame("whitespace_count")
ws_by_type_df["% of all whitespace"] = (ws_by_type / len(ws) * 100).round(1)
ws_by_type_df["% of this type that is whitespace"] = (
    ws_by_type / gdf["asset_sub_type"].value_counts() * 100
).round(1)
ws_by_type_df
```

```python
# Does missing last_inspection_date predict whitespace condition?

has_inspection = gdf["last_inspection_date"].notna()

inspection_vs_cond = pd.crosstab(
    has_inspection,
    gdf["_cond_status"],
)

inspection_vs_cond
```

```python


inspection_vs_cond["total"] = inspection_vs_cond.sum(axis=1)
inspection_vs_cond["whitespace_pct"] = (
    inspection_vs_cond["whitespace"] / inspection_vs_cond["total"] * 100
).round(1)
inspection_vs_cond
```

```python
# Maintainer breakdown for whitespace vs graded rows
maint_ct = pd.crosstab(gdf["asset_maintainer"], gdf["_cond_status"])
maint_ct["total"] = maint_ct.sum(axis=1)
maint_ct["whitespace_pct"] = (maint_ct["whitespace"] / maint_ct["total"] * 100).round(1)
maint_ct = maint_ct.sort_values("total", ascending=False)
maint_ct
```

```python
# How much of the whitespace is explained by Natural High Ground alone?
ws_nhg = ((gdf["_cond_status"] == "whitespace") & (gdf["asset_sub_type"] == "Natural High Ground")).sum()
print(f"Natural High Ground + whitespace: {ws_nhg:,} of {len(ws):,} whitespace rows ({ws_nhg/len(ws)*100:.1f}%)")

# How much by no inspection date?
ws_no_insp = ((gdf["_cond_status"] == "whitespace") & ~has_inspection).sum()
print(f"No inspection date + whitespace:  {ws_no_insp:,} of {len(ws):,} whitespace rows ({ws_no_insp/len(ws)*100:.1f}%)")

# Overlap: NHG AND no inspection date
ws_both = ((gdf["_cond_status"] == "whitespace") & (gdf["asset_sub_type"] == "Natural High Ground") & ~has_inspection).sum()
print(f"Both NHG + no inspection:         {ws_both:,} of {len(ws):,} whitespace rows ({ws_both/len(ws)*100:.1f}%)")

# Whitespace rows that are NOT Natural High Ground
ws_not_nhg = len(ws) - ws_nhg
print(f"\nWhitespace rows that are NOT Natural High Ground: {ws_not_nhg:,}")
if ws_not_nhg > 0:
    print("Their asset_sub_types:")
    print(ws[ws["asset_sub_type"] != "Natural High Ground"]["asset_sub_type"].value_counts())
```

## 6. Usable dataset after removing whitespace and null conditions

Drop all rows where `current_condition` is null or whitespace-only. What remains and what asset types does it contain?

```python
graded = gdf[gdf["_cond_status"].isin(["graded", "grade_0"])].copy()
excluded = gdf[~gdf["_cond_status"].isin(["graded", "grade_0"])]

print(f"Total rows:    {len(gdf):,}")
print(f"Excluded:      {len(excluded):,} (null or whitespace condition)")
print(f"Remaining:     {len(graded):,} (have a numeric condition grade 0-5)")
```

### Asset types in the graded vs excluded populations

For each asset sub-type: how many rows have a condition grade, how many don't, and what proportion is graded? This shows which asset types the condition grading system actually applies to.

```python
has_grade = gdf["_cond_status"].isin(["graded", "grade_0"])

type_summary = pd.DataFrame({
    "graded": gdf[has_grade].groupby("asset_sub_type").size(),
    "excluded": gdf[~has_grade].groupby("asset_sub_type").size(),
}).fillna(0).astype(int)

type_summary["total"] = type_summary["graded"] + type_summary["excluded"]
type_summary["graded_pct"] = (type_summary["graded"] / type_summary["total"] * 100).round(1)
type_summary = type_summary.sort_values("total", ascending=False)
type_summary
```

### Condition grade distribution within the graded subset

```python
graded_stripped = graded["current_condition"].str.strip()
grade_dist = graded_stripped.value_counts().sort_index().to_frame("count")
grade_dist["pct"] = (grade_dist["count"] / len(graded) * 100).round(2)
grade_dist
```

### Condition grade breakdown per asset type (graded rows only)

```python
grade_by_type = pd.crosstab(
    graded["asset_sub_type"],
    graded_stripped,
)
grade_by_type["total"] = grade_by_type.sum(axis=1)
grade_by_type = grade_by_type.sort_values("total", ascending=False)
grade_by_type
```

### Missingness in the graded subset

How does column coverage change once we restrict to rows that have a condition grade?

```python
n_graded = len(graded)
graded_miss = pd.DataFrame({
    "null_count": graded.drop(columns=["_cond_status"]).isnull().sum(),
    "coverage_pct": (graded.drop(columns=["_cond_status"]).notna().sum() / n_graded * 100).round(1),
    "full_dataset_coverage_pct": missingness["coverage_pct"],
}).sort_values("coverage_pct", ascending=True)

graded_miss["change"] = (graded_miss["coverage_pct"] - graded_miss["full_dataset_coverage_pct"]).round(1)
graded_miss
```
